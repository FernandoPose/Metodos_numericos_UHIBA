# Regresión lineal simple


La regresión lineal simple es una técnica fundamental en estadísticas y aprendizaje automático que se utiliza para modelar la relación entre una variable independiente (predictora) y una variable dependiente (objetivo) mediante una línea recta. Es una herramienta poderosa para entender y predecir relaciones lineales entre variables en el mundo real.


Variables en la Regresión Lineal Simple

Variable Independiente (X): También conocida como la variable predictora o explicativa. Es la variable que utilizamos para hacer predicciones.
Variable Dependiente (Y): Es la variable que queremos predecir o explicar a partir de la variable independiente.
La Ecuación de la Regresión Lineal Simple

La ecuación de la regresión lineal simple tiene la forma general: $ Y$ = $\beta_0$ +$\beta_1$ X +  ϵ

Donde:
$Y$ es la variable dependiente.
$X$ es la variable independiente.
$\beta_0$ es el intercepto o el valor de $Y$ cuando $X$ es igual a cero.
$\beta_1$ es la pendiente de la línea que representa la relación entre $X$ e $Y$.
$\epsilon$ representa el error residual, que es la diferencia entre la predicción y el valor real de $Y$.
Objetivo de la Regresión Lineal Simple

El objetivo es encontrar los valores de $\beta_0$ y $\beta_1$ que minimicen la suma de los cuadrados de los errores residuales, utilizando un método llamado "Mínimos Cuadrados".
Interpretación de $\beta_0$ y $\beta_1$

$\beta_0$: Representa el valor esperado de $Y$ cuando $X$ es igual a cero. Es el punto donde la línea de regresión corta el eje Y.

$\beta_1$: Representa el cambio esperado en $Y$ por cada unidad de cambio en $X$. Es la pendiente de la línea de regresión.

````python
# valores tomados de la clase
x_clase = np.array([1,2]).reshape((-1, 1))
y_clase = np.array([3,4])
````

# el porque de reshape
Cuando se utiliza la importación `from sklearn.linear_model import LinearRegression` en scikit-learn para crear un modelo de regresión lineal, a menudo se requiere que las características de entrada (`X`) se proporcionen en un formato específico. La razón detrás de la necesidad de darle esta forma particular a `X` (por ejemplo, `x_clase = np.array([1,2]).reshape((-1, 1))`) se relaciona con la estructura que espera el modelo de regresión lineal.

La razón detrás de esto se debe a que scikit-learn espera que las características de entrada `X` sean una matriz bidimensional, donde cada fila representa un ejemplo y cada columna representa una característica. Para la regresión lineal simple, donde tienes una sola característica predictora, necesitas darle a `X` la forma `(n_muestras, 1)`, donde `n_muestras` es el número de ejemplos o puntos de datos que tienes.

Aquí está la explicación paso a paso de la línea de código que proporcionaste:

1. `np.array([1,2])`: Esto crea un array de NumPy con dos elementos, que representan los valores de las características predictoras.

2. `.reshape((-1, 1))`: Esto es una operación de remodelado. El `-1` en la primera dimensión significa que NumPy debe inferir automáticamente el tamaño de esa dimensión para que coincida con el número de elementos en el array original. La segunda dimensión se establece en `1`, lo que significa que estamos transformando el array unidimensional en un array bidimensional con una sola columna.

El resultado es una matriz bidimensional `x_clase` que tiene la forma `(2, 1)`, lo que significa que tenemos dos ejemplos (filas) y una característica (columna). Esto es lo que scikit-learn espera cuando trabajas con regresión lineal simple.

En resumen, la transformación de la matriz de características `X` a la forma `(n_muestras, 1)` es una convención necesaria para que el modelo de regresión lineal simple en scikit-learn funcione correctamente. Esto asegura que los datos estén en el formato correcto para ser procesados por el modelo.

````python
fig, ax = plt.subplots(nrows=1, ncols=1,figsize=(10,6))
ax.scatter(x_clase, y_clase)
ax.grid(True)
ax.set_xlabel('x')
ax.set_ylabel('y')
#ax.legend()
plt.show()
````

## Entrenamiento 💪

````python
modelo_clase = LinearRegression().fit(x_clase, y_clase)
````

````python
print('beta0:', modelo_clase.intercept_)
print('beta1: ', modelo_clase.coef_[0])
````

````python
# Podemos predecir usando el modelo_clase
y_pred_clase = modelo_clase.predict(x_clase)
# Por último, calculamos el error cuadrático med ,dístico R^2
print('Error cuadrático medio: ', mean_squared_error(y_clase, y_pred_clase))
print('R_2: ',r2_score(y_clase, y_pred_clase))
````

El **Error Cuadrático Medio** (MSE, por sus siglas en inglés, Mean Squared Error) es una métrica comúnmente utilizada en el aprendizaje automático para evaluar la calidad de un modelo de regresión. El MSE mide la diferencia entre las predicciones de un modelo y los valores reales de los datos. Cuanto menor sea el MSE, mejor será el rendimiento del modelo.

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Datos de Pedro ajustados
data_pedro_ajustado = {
    "Año": [1, 2, 4, 5, 6, 7],
    "Exámenes Finales Aprobados": [3, 5, 9, 10, 8, 9]
}
df_pedro_ajustado = pd.DataFrame(data_pedro_ajustado)

# Crear el modelo de regresión lineal
X = df_pedro_ajustado["Año"].values.reshape(-1, 1)
y = df_pedro_ajustado["Exámenes Finales Aprobados"].values

# Ajuste del modelo original
model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)

# Generar varias líneas de regresión con diferentes pendientes y diferentes colores
plt.figure(figsize=(10, 6))

# Graficar los puntos de Pedro
plt.scatter(df_pedro_ajustado["Año"], df_pedro_ajustado["Exámenes Finales Aprobados"], color='black', label="Datos de Pedro")

# Línea de regresión original (azul)
plt.plot(df_pedro_ajustado["Año"], y_pred, color='blue', label=f"Línea de Regresión Azul")

# Generar líneas adicionales con diferentes pendientes
for factor, color in zip([0.5, 1.5, 2.0], ['magenta', 'green', 'red']):
    adjusted_model = LinearRegression()
    adjusted_model.fit(X, y * factor)
    y_adjusted_pred = adjusted_model.predict(X)
    plt.plot(df_pedro_ajustado["Año"], y_adjusted_pred, color=color, label=f"Línea {color}")

# Título y etiquetas
plt.title("Regresión Lineal con Varias Líneas", fontsize=14)
plt.xlabel("Año", fontsize=12)
plt.ylabel("Exámenes Finales Aprobados", fontsize=12)
plt.grid(True)
plt.legend()

# Mostrar la gráfica
plt.show()
````

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Datos de Pedro ajustados
data_pedro_ajustado = {
    "Año": [1, 2, 4, 5, 6, 7],
    "Exámenes Finales Aprobados": [3, 5, 9, 10, 8, 9]
}
df_pedro_ajustado = pd.DataFrame(data_pedro_ajustado)

# Crear el modelo de regresión lineal
X = df_pedro_ajustado["Año"].values.reshape(-1, 1)
y = df_pedro_ajustado["Exámenes Finales Aprobados"].values

# Ajuste del modelo original
model = LinearRegression()
model.fit(X, y)

# Predicción para el valor de x = 3
y_pred_3 = model.predict([[3]])

# Predicción de todos los valores para dibujar la línea
y_pred = model.predict(X)

# Graficar los puntos originales de Pedro
plt.figure(figsize=(10, 6))
plt.scatter(df_pedro_ajustado["Año"], df_pedro_ajustado["Exámenes Finales Aprobados"], color='black', label="Datos Originales")

# Graficar la línea de regresión
plt.plot(df_pedro_ajustado["Año"], y_pred, color='blue', label="Línea de Regresión")

# Marcar la predicción para x = 3
plt.scatter(3, y_pred_3, color='red', label=f'Predicción en x=3 ({y_pred_3[0]:.2f})', s=100, marker='x')

# Marcar el valor original en x = 3 (sabemos que es 6)
plt.scatter(3, 6, color='green', label='Valor Perdido en x=3 (6)', s=100, marker='o')

# Título y etiquetas
plt.title("Regresión Lineal - Predicción en x=3 vs Valor Perdido", fontsize=14)
plt.xlabel("Año", fontsize=12)
plt.ylabel("Exámenes Finales Aprobados", fontsize=12)
plt.grid(True)
plt.legend()

# Mostrar la gráfica
plt.show()
````
