# Regresin lineal simple


La regresin lineal simple es una tcnica fundamental en estadsticas y aprendizaje automtico que se utiliza para modelar la relacin entre una variable independiente (predictora) y una variable dependiente (objetivo) mediante una lnea recta. Es una herramienta poderosa para entender y predecir relaciones lineales entre variables en el mundo real.


Variables en la Regresin Lineal Simple

Variable Independiente (X): Tambin conocida como la variable predictora o explicativa. Es la variable que utilizamos para hacer predicciones.
Variable Dependiente (Y): Es la variable que queremos predecir o explicar a partir de la variable independiente.
La Ecuacin de la Regresin Lineal Simple

La ecuacin de la regresin lineal simple tiene la forma general: $ Y$ = $\beta_0$ +$\beta_1$ X +  

Donde:
$Y$ es la variable dependiente.
$X$ es la variable independiente.
$\beta_0$ es el intercepto o el valor de $Y$ cuando $X$ es igual a cero.
$\beta_1$ es la pendiente de la lnea que representa la relacin entre $X$ e $Y$.
$\epsilon$ representa el error residual, que es la diferencia entre la prediccin y el valor real de $Y$.
Objetivo de la Regresin Lineal Simple

El objetivo es encontrar los valores de $\beta_0$ y $\beta_1$ que minimicen la suma de los cuadrados de los errores residuales, utilizando un mtodo llamado "Mnimos Cuadrados".
Interpretacin de $\beta_0$ y $\beta_1$

$\beta_0$: Representa el valor esperado de $Y$ cuando $X$ es igual a cero. Es el punto donde la lnea de regresin corta el eje Y.

$\beta_1$: Representa el cambio esperado en $Y$ por cada unidad de cambio en $X$. Es la pendiente de la lnea de regresin.

````python
# valores tomados de la clase
x_clase = np.array([1,2]).reshape((-1, 1))
y_clase = np.array([3,4])
````

# el porque de reshape
Cuando se utiliza la importacin `from sklearn.linear_model import LinearRegression` en scikit-learn para crear un modelo de regresin lineal, a menudo se requiere que las caractersticas de entrada (`X`) se proporcionen en un formato especfico. La razn detrs de la necesidad de darle esta forma particular a `X` (por ejemplo, `x_clase = np.array([1,2]).reshape((-1, 1))`) se relaciona con la estructura que espera el modelo de regresin lineal.

La razn detrs de esto se debe a que scikit-learn espera que las caractersticas de entrada `X` sean una matriz bidimensional, donde cada fila representa un ejemplo y cada columna representa una caracterstica. Para la regresin lineal simple, donde tienes una sola caracterstica predictora, necesitas darle a `X` la forma `(n_muestras, 1)`, donde `n_muestras` es el nmero de ejemplos o puntos de datos que tienes.

Aqu est la explicacin paso a paso de la lnea de cdigo que proporcionaste:

1. `np.array([1,2])`: Esto crea un array de NumPy con dos elementos, que representan los valores de las caractersticas predictoras.

2. `.reshape((-1, 1))`: Esto es una operacin de remodelado. El `-1` en la primera dimensin significa que NumPy debe inferir automticamente el tamao de esa dimensin para que coincida con el nmero de elementos en el array original. La segunda dimensin se establece en `1`, lo que significa que estamos transformando el array unidimensional en un array bidimensional con una sola columna.

El resultado es una matriz bidimensional `x_clase` que tiene la forma `(2, 1)`, lo que significa que tenemos dos ejemplos (filas) y una caracterstica (columna). Esto es lo que scikit-learn espera cuando trabajas con regresin lineal simple.

En resumen, la transformacin de la matriz de caractersticas `X` a la forma `(n_muestras, 1)` es una convencin necesaria para que el modelo de regresin lineal simple en scikit-learn funcione correctamente. Esto asegura que los datos estn en el formato correcto para ser procesados por el modelo.

````python
fig, ax = plt.subplots(nrows=1, ncols=1,figsize=(10,6))
ax.scatter(x_clase, y_clase)
ax.grid(True)
ax.set_xlabel('x')
ax.set_ylabel('y')
#ax.legend()
plt.show()
````

## Entrenamiento 

````python
modelo_clase = LinearRegression().fit(x_clase, y_clase)
````

````python
print('beta0:', modelo_clase.intercept_)
print('beta1: ', modelo_clase.coef_[0])
````

````python
# Podemos predecir usando el modelo_clase
y_pred_clase = modelo_clase.predict(x_clase)
# Por ltimo, calculamos el error cuadrtico med ,dstico R^2
print('Error cuadrtico medio: ', mean_squared_error(y_clase, y_pred_clase))
print('R_2: ',r2_score(y_clase, y_pred_clase))
````

El **Error Cuadrtico Medio** (MSE, por sus siglas en ingls, Mean Squared Error) es una mtrica comnmente utilizada en el aprendizaje automtico para evaluar la calidad de un modelo de regresin. El MSE mide la diferencia entre las predicciones de un modelo y los valores reales de los datos. Cuanto menor sea el MSE, mejor ser el rendimiento del modelo.

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Datos de Pedro ajustados
data_pedro_ajustado = {
    "Ao": [1, 2, 4, 5, 6, 7],
    "Exmenes Finales Aprobados": [3, 5, 9, 10, 8, 9]
}
df_pedro_ajustado = pd.DataFrame(data_pedro_ajustado)

# Crear el modelo de regresin lineal
X = df_pedro_ajustado["Ao"].values.reshape(-1, 1)
y = df_pedro_ajustado["Exmenes Finales Aprobados"].values

# Ajuste del modelo original
model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)

# Generar varias lneas de regresin con diferentes pendientes y diferentes colores
plt.figure(figsize=(10, 6))

# Graficar los puntos de Pedro
plt.scatter(df_pedro_ajustado["Ao"], df_pedro_ajustado["Exmenes Finales Aprobados"], color='black', label="Datos de Pedro")

# Lnea de regresin original (azul)
plt.plot(df_pedro_ajustado["Ao"], y_pred, color='blue', label=f"Lnea de Regresin Azul")

# Generar lneas adicionales con diferentes pendientes
for factor, color in zip([0.5, 1.5, 2.0], ['magenta', 'green', 'red']):
    adjusted_model = LinearRegression()
    adjusted_model.fit(X, y * factor)
    y_adjusted_pred = adjusted_model.predict(X)
    plt.plot(df_pedro_ajustado["Ao"], y_adjusted_pred, color=color, label=f"Lnea {color}")

# Ttulo y etiquetas
plt.title("Regresin Lineal con Varias Lneas", fontsize=14)
plt.xlabel("Ao", fontsize=12)
plt.ylabel("Exmenes Finales Aprobados", fontsize=12)
plt.grid(True)
plt.legend()

# Mostrar la grfica
plt.show()
````
