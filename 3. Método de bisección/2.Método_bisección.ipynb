
Para la memoria 🧠
`np.linspace` es una función proporcionada por la biblioteca NumPy que devuelve un array de números equiespaciados en un intervalo especificado. La función es muy útil para evaluar funciones en un rango o para generar puntos en un intervalo.

La sintaxis básica de `np.linspace` es:


`np.linspace(start, stop, num=50)`


Donde:
- `start` es el valor inicial del intervalo.
- `stop` es el valor final del intervalo.
- `num` es el número de puntos que se quieren generar. Por defecto, es 50.

Entonces, `np.linspace` genera `num` puntos equiespaciados entre `start` y `stop`, inclusive.

Por ejemplo, si ejecutas `np.linspace(0, 1, 5)`, obtendrás un array con los valores `[0, 0.25, 0.5, 0.75, 1]`.

## Encontrar raices Reales🌱 🔍


```python
import numpy as np
def raices_cuadraticas(a, b, c):
    """
    Encuentra las raíces de la ecuación cuadrática ax^2 + bx + c = 0.

    Parámetros:
        a, b, c: coeficientes de la ecuación cuadrática.

    Retorna:
        raices: una tupla con las dos raíces (pueden ser reales o complejas).
    """

    discriminante = b**2 - 4*a*c

    if discriminante > 0:  # Dos raíces reales y distintas
        x1 = (-b + np.sqrt(discriminante)) / (2*a)
        x2 = (-b - np.sqrt(discriminante)) / (2*a)
        return (x1, x2)
    elif discriminante == 0:  # Una raíz real y doble
        x1 = -b / (2*a)
        return (x1,)
    else:  # Dos raíces complejas
        real = -b / (2*a)
        imag = np.sqrt(-discriminante) / (2*a)
        return (complex(real, imag), complex(real, -imag))

#raices = raices_cuadraticas(1, 0, -4)

```

```python
# a*x**2+b*x+c
raices_cuadraticas(1, 0, -4)
#help(raices_cuadraticas)
```

# Método de la Bisección:
Este es un método de búsqueda incremental. Si conoces que una función cambia de signo en un intervalo, puedes dividir ese intervalo por la mitad **repetidamente** hasta que encuentres una aproximación de la raíz.


La idea básica del método de la bisección para encontrar raíces de una función es la siguiente:




imagen.png


```python
# Marcar puntos
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(8,6))
x = np.linspace(0,8.5,105)
y = funcion_cuadratica(x)
ax.plot(x, y, label=r'$f(x) = x^2 - 4$', color='grey')
ax.axhline(0, color='black',linewidth=1.5)
ax.axvline(0, color='black',linewidth=1.5)
#------------------------------

x_a = 0
f_a = funcion_cuadratica(x_a)
ax.text(x_a,f_a,"f(a)", color = 'darkmagenta')
ax.scatter(x_a,f_a,s=35, color='darkmagenta',marker='s')

x_b = 8
f_b = funcion_cuadratica(x_b)
ax.text(x_b,f_b,"f(b)", color = 'darkmagenta')
ax.scatter(x_b,f_b,s=35, color='darkmagenta',marker='s')

x_c = valor_medio(x_b,x_a)
f_c = funcion_cuadratica(x_c)
ax.text(x_c,f_c,"f(c)", color = 'peru')
ax.scatter(x_c,f_c,s=35, color='peru',marker='s')


x_ini =x_a
x_fin = x_b
y_flecha = 70
ax.annotate('', xy=(x_ini, y_flecha),
                   xytext=(x_fin, y_flecha),
                arrowprops=dict(arrowstyle='<->', color='green'))
ax.text(x_ini,y_flecha,"a", color = 'peru')
ax.text(x_fin,y_flecha,"b", color = 'peru')
#-------------------------------
ax.set_xlabel('x')
ax.set_ylabel('f(x)')
ax.set_title(r'Gráfica de $f(x) = x^2 - 4$ raíz x > 0')
ax.set_ylim([-5,85])
ax.grid(True, linestyle='--', linewidth=0.5)
ax.legend()
plt.show()
```
