# Errores absoluto y relativo [Queda para el Lunes ?]
Es crucial cuantificar la precisin de una aproximacin. Los errores absoluto y relativo son dos medidas comunes para hacerlo




## **Error Absoluto**:
El error absoluto se define como la diferencia entre el valor verdadero (o de referencia) y el valor aproximado.

\begin{equation}
\text{Error Absoluto} = | \text{Valor Verdadero} - \text{Valor Aproximado} |
\end{equation}

## **Error Relativo**:
El error relativo es el error absoluto dividido por el valor absoluto del valor verdadero. Se puede expresar como:

\begin{equation}
\text{Error Relativo} = \frac{\text{Error Absoluto}}{|\text{Valor Verdadero}|}
\end{equation}


Nota: para expresarlo en porcentaje, simplemente multiplicas el error relativo por 100.

````python
import math
import matplotlib.pyplot as plt

# Funcin para calcular el rea del crculo
def area_circulo(r, pi_value):
    return pi_value * r**2

# Valores de 
valor_verdadero_pi = math.pi
aproximacion_pi = 3

# Conjunto de radios
radios = list(range(1, 21))

# Calcula las reas usando el valor verdadero de 
areas_verdadero = [area_circulo(r, valor_verdadero_pi) for r in radios]

# Calcula las reas usando la aproximacin de 
areas_aproximado = [area_circulo(r, aproximacion_pi) for r in radios]

# Graficar
plt.figure(figsize=(10, 6))
plt.plot(radios, areas_verdadero, label='rea con  verdadero', marker='o')
plt.plot(radios, areas_aproximado, label='rea con  aproximado', marker='x')
plt.xlabel('Radio')
plt.ylabel('rea')
plt.title('rea del crculo: Valor verdadero vs. Aproximacin')
plt.legend()
plt.grid(True)
plt.show()
````
