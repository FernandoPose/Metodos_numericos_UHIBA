
Para la memoria 游
`np.linspace` es una funci칩n proporcionada por la biblioteca NumPy que devuelve un array de n칰meros equiespaciados en un intervalo especificado. La funci칩n es muy 칰til para evaluar funciones en un rango o para generar puntos en un intervalo.

La sintaxis b치sica de `np.linspace` es:


`np.linspace(start, stop, num=50)`


Donde:
- `start` es el valor inicial del intervalo.
- `stop` es el valor final del intervalo.
- `num` es el n칰mero de puntos que se quieren generar. Por defecto, es 50.

Entonces, `np.linspace` genera `num` puntos equiespaciados entre `start` y `stop`, inclusive.

Por ejemplo, si ejecutas `np.linspace(0, 1, 5)`, obtendr치s un array con los valores `[0, 0.25, 0.5, 0.75, 1]`.

## Encontrar raices Reales游꺔 游댌


```python
import numpy as np
def raices_cuadraticas(a, b, c):
    """
    Encuentra las ra칤ces de la ecuaci칩n cuadr치tica ax^2 + bx + c = 0.

    Par치metros:
        a, b, c: coeficientes de la ecuaci칩n cuadr치tica.

    Retorna:
        raices: una tupla con las dos ra칤ces (pueden ser reales o complejas).
    """

    discriminante = b**2 - 4*a*c

    if discriminante > 0:  # Dos ra칤ces reales y distintas
        x1 = (-b + np.sqrt(discriminante)) / (2*a)
        x2 = (-b - np.sqrt(discriminante)) / (2*a)
        return (x1, x2)
    elif discriminante == 0:  # Una ra칤z real y doble
        x1 = -b / (2*a)
        return (x1,)
    else:  # Dos ra칤ces complejas
        real = -b / (2*a)
        imag = np.sqrt(-discriminante) / (2*a)
        return (complex(real, imag), complex(real, -imag))

#raices = raices_cuadraticas(1, 0, -4)

```

```python
# a*x**2+b*x+c
raices_cuadraticas(1, 0, -4)
#help(raices_cuadraticas)
```

# M칠todo de la Bisecci칩n:
Este es un m칠todo de b칰squeda incremental. Si conoces que una funci칩n cambia de signo en un intervalo, puedes dividir ese intervalo por la mitad **repetidamente** hasta que encuentres una aproximaci칩n de la ra칤z.


La idea b치sica del m칠todo de la bisecci칩n para encontrar ra칤ces de una funci칩n es la siguiente:




imagen.png


```python
# Marcar puntos
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(8,6))
x = np.linspace(0,8.5,105)
y = funcion_cuadratica(x)
ax.plot(x, y, label=r'$f(x) = x^2 - 4$', color='grey')
ax.axhline(0, color='black',linewidth=1.5)
ax.axvline(0, color='black',linewidth=1.5)
#------------------------------

x_a = 0
f_a = funcion_cuadratica(x_a)
ax.text(x_a,f_a,"f(a)", color = 'darkmagenta')
ax.scatter(x_a,f_a,s=35, color='darkmagenta',marker='s')

x_b = 8
f_b = funcion_cuadratica(x_b)
ax.text(x_b,f_b,"f(b)", color = 'darkmagenta')
ax.scatter(x_b,f_b,s=35, color='darkmagenta',marker='s')

x_c = valor_medio(x_b,x_a)
f_c = funcion_cuadratica(x_c)
ax.text(x_c,f_c,"f(c)", color = 'peru')
ax.scatter(x_c,f_c,s=35, color='peru',marker='s')


x_ini =x_a
x_fin = x_b
y_flecha = 70
ax.annotate('', xy=(x_ini, y_flecha),
                   xytext=(x_fin, y_flecha),
                arrowprops=dict(arrowstyle='<->', color='green'))
ax.text(x_ini,y_flecha,"a", color = 'peru')
ax.text(x_fin,y_flecha,"b", color = 'peru')
#-------------------------------
ax.set_xlabel('x')
ax.set_ylabel('f(x)')
ax.set_title(r'Gr치fica de $f(x) = x^2 - 4$ ra칤z x > 0')
ax.set_ylim([-5,85])
ax.grid(True, linestyle='--', linewidth=0.5)
ax.legend()
plt.show()
```
