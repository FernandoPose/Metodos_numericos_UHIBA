```python
# Definimos los parmetros
N = 3
b= 30
a= 0
h = (b-a)/N  # paso calculado
mu = 0.05  # valor de mu

# Condiciones iniciales
x = [0]
y = [200]

# Funcin derivada
def df(x, y):
    return mu * y

# Bucle de Euler
for i in range(N):
    xmas1 = x[i] + h
    x.append(xmas1)
    ymas1 = y[i] + h * df(x[i], y[i])
    y.append(ymas1)

print("x:", x)
print("y:", y)
```

```python
import pandas as pd

# Convertir las listas a un DataFrame
df_resultado = pd.DataFrame({
    'xi': x,
    'yi': y
})

df_resultado
```

```python
import pandas as pd

# Definimos los parmetros
N = 3
b= 30
a= 0
h = (b-a)/N  # paso calculado
mu = 0.05  # valor de mu

# Condiciones iniciales
x = [0]
y = [200]

# Funcin derivada
def f(x, y):
    return mu * y

# Bucle de Euler modificado
for i in range(N):
    # Primer paso - Calcular m1
    m1 = f(x[i], y[i])

    # Segundo paso - Prediccin provisional de y_{i+1}
    y_provisional = y[i] + h * m1

    # Tercer paso - Calcular m2
    m2 = f(x[i] + h, y_provisional)

    # Promedio de pendientes
    phi = (m1 + m2) / 2

    # Actualizar x e y
    xmas1 = x[i] + h
    x.append(xmas1)

    ymas1 = y[i] + h * phi
    y.append(ymas1)

# Convertir a DataFrame
df_resultado = pd.DataFrame({
    'xi': x,
    'yi': y
})

print(df_resultado)
```
