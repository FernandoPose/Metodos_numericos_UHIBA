# Errores absoluto y relativo
Es crucial cuantificar la precisión de una aproximación. Los errores absoluto y relativo son dos medidas comunes para hacerlo




## **Error Absoluto**:
El error absoluto se define como la diferencia entre el valor verdadero (o de referencia) y el valor aproximado.

\begin{equation}
\text{Error Absoluto} = | \text{Valor Verdadero} - \text{Valor Aproximado} |
\end{equation}

## **Error Relativo**:
El error relativo es el error absoluto dividido por el valor absoluto del valor verdadero. Se puede expresar como:

\begin{equation}
\text{Error Relativo} = \frac{\text{Error Absoluto}}{|\text{Valor Verdadero}|}
\end{equation}


Nota: para expresarlo en porcentaje, simplemente multiplicas el error relativo por 100.

````python
def errores(vv, va):
  err_abs = round(abs(vv-va),5)
  err_rel = round(err_abs/round(vv),5)
  print("El error absoluto es: ", err_abs)
  print("El error relativo es: ", err_rel)
  return err_abs, err_rel
````

# Ejercicio: Cálculo del área de un círculo

Calcular y comparar gráficamente el área de un círculo utilizando el valor verdadero de π (math.pi , no te olvides de importar la lib)y su aproximación (3.14).

1. los radios varian de 1 a 10
2. crea y utiliza una función A=πr^2
3.  guarda en variables distintas los valores verdaderos y el aproximado
4. Grafica las áreas calculadas con el valor verdadero vs. la aproximación para cada radio (no te olvides de importar matplotlib)
5. Grafica los errores_abs y errores_relativos en otro gráfico
  

````python
import math
import matplotlib.pyplot as plt

# Función para calcular el área del círculo
def area_circulo(r, pi_value):
    return pi_value * r**2

# Valores de π
valor_verdadero_pi = math.pi
aproximacion_pi = 3

# Conjunto de radios
radios = list(range(1, 11))

# Calcula las áreas usando el valor verdadero de π
areas_verdadero = [area_circulo(r, valor_verdadero_pi) for r in radios]

# Calcula las áreas usando la aproximación de π
areas_aproximado = [area_circulo(r, aproximacion_pi) for r in radios]

# Graficar
plt.figure(figsize=(10, 6))
plt.plot(radios, areas_verdadero, label='Área con π verdadero', marker='o')
plt.plot(radios, areas_aproximado, label='Área con π aproximado', marker='x')
plt.xlabel('Radio')
plt.ylabel('Área')
plt.title('Área del círculo: Valor verdadero vs. Aproximación')
plt.legend()
plt.grid(True)
plt.show()

#Mismo proceso para errores
errores_abs = [abs(av-aa) for av,aa in zip(areas_verdadero,areas_aproximado)]
print(errores_abs)

errores_rel = [ea/abs(av) for ea,av in zip(errores_abs,areas_verdadero)]
print(errores_rel)
````

````python
#Graficar errores absolutos
plt.figure(figsize=(10, 6))
plt.plot(radios, errores_abs, label='Error absoluto', marker='o')
plt.xlabel('Radio')
plt.ylabel('Error Absoluto')
plt.title('Área del círculo: Error Absoluto')
plt.legend()
plt.grid(True)
plt.show()
````

````python
#Graficar errores relativos
plt.figure(figsize=(10, 6))
plt.plot(radios, errores_rel, label='Error relativo', marker='x')
plt.xlabel('Radio')
plt.ylabel('Error Relativo')
plt.title('Área del círculo: Error Relativo')
plt.legend()
plt.grid(True)
plt.show()
````

````python
#Graficar ambos errores juntos
plt.figure(figsize=(10, 6))
plt.plot(radios, errores_abs, label='Error absoluto', marker='o')
plt.plot(radios, errores_rel, label='Error relativo', marker='x')
plt.xlabel('Radio')
plt.ylabel('Errores')
plt.title('Área del círculo: Errores Absoluto y Relativo')
plt.legend()
plt.grid(True)
plt.show()
````

# Ejercicio: Cálculo de Errores para la Función Exponencial

 Calcular los errores absoluto y relativo al evaluar una aproximación de la función exponencial para una lista de valores de `x`.

 1. Considera la siguiente lista de valores para x: [−1,−0.5,0,0.5,1,5,10].
 2. Para cada valor de `x`, calcula:
  * El valor exacto de `e^x ` utilizando `math.exp()`
  * Una aproximación de `e^x ` utilizando la serie de Taylor truncada hasta el término
`x^3`
\begin{equation}
e^x \approx 1 + x + \frac{x^2}{2!} + \frac{x^3}{3!}
\end{equation}

3. Una vez obtenidos el valor exacto y la aproximación para cada `x`, calcula el error absoluto y el error relativo.

4. En una plot, grafica x, valor_exacto , valor_aproximado. Si puedes coloca las legends para identificar

5. Guarda todo lo que creas necesario en un dataframe



Para x_valores = [-1, -0.5, 0, 0.5, 1] el resultao esperado es Sin título.png **¿qué sucede en x  5 y 10?**

````python
import math
import matplotlib.pyplot as plt
import pandas as pd

x_val_1 = [-1,-0.5,0,0.5,1,5,10]
x_val_2 = [-1,-0.5,0,0.5,1]

def exp_aprox(n):
  return 1 + n + n**2/math.factorial(2) + n**3/math.factorial(3)

#Lista con valores verdaderos
exp_verdadero = [math.exp(i) for i in x_val_1]

#Lista con valores aproximados con serie de Taylor
exp_taylor = [exp_aprox(i) for i in x_val_1]

#Lista con errores absolutos
errores_abs = [abs(av-aa) for av,aa in zip(exp_verdadero,exp_taylor)]

#Lista con errores relativos
errores_rel = [ea/abs(av) for ea,av in zip(errores_abs,exp_verdadero)]

#Guardar datos en dataframe
datos = {'x':x_val_1,
         'e^x':exp_verdadero,
         'e^x_aproximado':exp_taylor,
         'error_absoluto':errores_abs,
         'error_relativo':errores_rel}

df = pd.DataFrame(datos)


df
````

````python
#Graficar para x:[-1,1]
plt.figure(figsize=(10, 6))
plt.plot(df['x'][:5], df['e^x'][:5], label='Exponencial verdadera', marker='o')
plt.plot(df['x'][:5], df['e^x_aproximado'][:5], label='Exponencial aproximada', marker='x')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Exponencial: Valor verdadero vs. Aproximación')
plt.legend()
plt.grid(True)
plt.show()
````

````python
#Graficar para x:[-1,10]
plt.figure(figsize=(10, 6))
plt.plot(df['x'], df['e^x'], label='Exponencial verdadera', marker='o')
plt.plot(df['x'], df['e^x_aproximado'], label='Exponencial aproximada', marker='x')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Exponencial: Valor verdadero vs. Aproximación')
plt.legend()
plt.grid(True)
plt.show()
````
